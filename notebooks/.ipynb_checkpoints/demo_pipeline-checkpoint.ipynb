```python
from image_model import ImageModel
import torch

model = ImageModel()

image_features_list = []

for i in range(len(df)):
    _, image, _, _ = load_sample(df, i)

    image = image.unsqueeze(0)  # add batch dimension

    with torch.no_grad():
        features = model(image)

    image_features_list.append(features.squeeze().numpy())

print(len(image_features_list))
```

```python
import numpy as np

X_image = np.array(image_features_list)

print(X_image.shape)
```

```python
y = df["label"].values

print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# split data
X_train, X_test, y_train, y_test = train_test_split(
    X_image, y, test_size=0.25, random_state=42
)

# train model
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

# predict
y_pred = clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
metadata_list = []

for i in range(len(df)):
    _, _, metadata, _ = load_sample(df, i)
    metadata_list.append(metadata)

X_meta = np.array(metadata_list)

print(X_meta.shape)
```

```python
X_combined = np.concatenate((X_image, X_meta), axis=1)

print(X_combined.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# split
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.25, random_state=42
)

# train
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

# predict
y_pred = clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
